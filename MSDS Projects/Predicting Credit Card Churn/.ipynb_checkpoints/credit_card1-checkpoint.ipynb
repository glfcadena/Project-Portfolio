### Data Cleaning and Pre-Processing (Step 1 to 4)

#### Step 1. Identify missing values in tabular data
Missing values were identified in the bar plots under Exploratory Data Analysis where they were tagged as `Unknown` datapoints. 

#### Step 2. Eliminate samples or features with missing values
The missing values tagged as 'Unknown' are removed in the dataset before feeding it into the machine learning model. A total of 3,380 unknown values would be removed.

```python
# Dropping Unknown in the Dataset
bank_df = df[df.Marital_Status != 'Unknown']
bank_df = bank_df[bank_df.Education_Level != 'Unknown']
bank_df = bank_df[bank_df.Income_Category != 'Unknown']
```

#### Step 3. Imputing missing values
No missing values were imputed since they were removed.

#### Step 4. Handling Categorical data
Ordinal features like `Education_Level` and `Card_Category` were converted to numerical values but still represented ranking. Nominal features like `Attrition_Flag` and `Gender` were converted to binary representative (1,0) and nominal features like `Marital_Status` and `Income_Category` were one-hot-encoded to represent their numerical values.

```python
# Mapping Ordinal Variables
education = {
    'Uneducated': 0,
    'High School': 1,
    'College': 2,
    'Post-Graduate': 3,
    'Doctorate': 4
}

card = {
    'Blue': 0,
    'Silver': 1,
    'Gold': 2,
    'Platinum': 3
}



# Converting Categorical Variables to Numerical
bank_df.Attrition_Flag = np.where(bank_df.Attrition_Flag == 'Existing Customer', 0, 1)
bank_df.Gender = np.where(bank_df.Gender == 'Male', 1, 0)
bank_df.Education_Level = np.where(bank_df.Education_Level == 'Graduate', 'College', bank_df.Education_Level)
bank_df.Education_Level = bank_df.Education_Level.map(education)
bank_df.Card_Category = bank_df.Card_Category.map(card)

# One-Hot Encoding of Categorical Variables
bank_df = pd.get_dummies(bank_df, columns=['Marital_Status'], drop_first=True)

from sklearn.preprocessing import OrdinalEncoder

# Use Ordinal Encoder for education level, income status, and card category
income_map = ['Less than $40K', '$40K - $60K', '$60K - $80K',  '$80K - $120K', '$120K +']
oe = OrdinalEncoder(categories=[income_map])
bank_df['Income_Category'] = oe.fit_transform(bank_df[['Income_Category']])
```

### automl results using default hyperparameters of sklearn classifiers, use an appropriate metric beginning at this step. Include a brief (1-2 sentences) discussion of the chosen metric.


Chosen metric is recall since there is a higher cost associated with attrited customers than those that are existing

```python
skf = StratifiedKFold(n_splits=5)

res = {}

# log start time
total_start = time.time()
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
index = 0
for model_name, model in tqdm(models_dict.items()):
    train_scores = []
    val_scores = []
    
    for train_index, val_index in skf.split(X_trainval, y_trainval):
        X_train, X_val = X_trainval.iloc[train_index], X_trainval.iloc[val_index]
        y_train, y_val = y_trainval.iloc[train_index], y_trainval.iloc[val_index]

        start_time = time.time()

        # fit
        model.fit(X_train, y_train)

        train_preds = model.predict(X_train)
        val_preds = model.predict(X_val)

        train_score = recall_score(y_train, train_preds)
        val_score = recall_score(y_val, val_preds)

        end_time = time.time()

        train_scores.append(train_score)
        val_scores.append(val_score)
        
        res[model_name] = {
            'ave_train_recall':np.mean(train_scores) * 100,
            'ave_val_recall':np.mean(val_scores) * 100,
            'run_time': end_time - start_time
        }
        
    plot_confusion_matrix(model, X_val, y_val,
                         ax=axes[index], cmap='summer',
                         )
    axes[index].set_title(model_name, fontsize=12)
    axes[index].set_xlabel('Predicted Label', fontsize=12)
    axes[index].set_ylabel('True Label', fontsize=12)
    index += 1

# log end time
total_end = time.time()

elapsed = total_end - total_start
print(f"Report Generated in {elapsed:.2f} seconds")

res = pd.DataFrame(res).T
display(res)
```

```python
skf = StratifiedKFold(n_splits=5)
cats = [1, 2, 3, 4, 5, 18, 19]
res_methods = [SMOTE(sampling_strategy=0.4, n_jobs=-1, random_state=143),
               ADASYN(sampling_strategy=0.4, n_jobs=-1, random_state=143), 
               SMOTENC(sampling_strategy=0.4, n_jobs=-1, random_state=143, 
                       categorical_features=cats)]

res_names = ['SMOTE', 'ADASYN', 'SMOTENC']
res = {}

# log start time
total_start = time.time()

fig, axes = plt.subplots(3, 3, figsize=(20, 20))
axes = axes.flatten()
index = 0
for model_name, model in tqdm(models_dict.items()):
    train_scores = []
    val_scores = []
    for res_name, resample in zip(res_names, res_methods):
        start_time = time.time()    
        for train_index, val_index in skf.split(X_trainval, y_trainval):
            X_train, X_val = X_trainval.iloc[train_index], X_trainval.iloc[val_index]
            y_train, y_val = y_trainval.iloc[train_index], y_trainval.iloc[val_index]

            X_train, y_train = resample.fit_resample(X_train, y_train)
            # fit
            model.fit(X_train, y_train)

            train_preds = model.predict(X_train)
            val_preds = model.predict(X_val)

            train_score = recall_score(y_train, train_preds)
            val_score = recall_score(y_val, val_preds)

        

            train_scores.append(train_score)
            val_scores.append(val_score)
        
        end_time = time.time()
        res[(model_name, res_name)] = {
            'ave_train_recall':np.mean(train_scores) * 100,
            'ave_val_recall':np.mean(val_scores) * 100,
            'run_time': end_time - start_time
        }

        plot_confusion_matrix(model, X_val, y_val,
                            ax=axes[index], cmap='summer',
                            )
        axes[index].set_title(f'{model_name} - {res_name}', fontsize=12)
        axes[index].set_xlabel('Predicted Label', fontsize=12)
        axes[index].set_ylabel('True Label', fontsize=12)
        index += 1

# log end time
total_end = time.time()

elapsed = total_end - total_start
print(f"Report Generated in {elapsed:.2f} seconds")

res = pd.DataFrame(res).T
display(res)
```

### Sensitivity Analysis 

#### SMOTE

```python
skf = StratifiedKFold(n_splits=5)
model = GradientBoostingClassifier(random_state=11) 
res_smote = {}
for k in range(1, 8):
    train_scores = []
    val_scores = []
    start_time = time.time()
    for train_index, val_index in skf.split(X_trainval, y_trainval):
        X_train, X_val = X_trainval.iloc[train_index], X_trainval.iloc[val_index]
        y_train, y_val = y_trainval.iloc[train_index], y_trainval.iloc[val_index]

        resample = SMOTE(sampling_strategy=0.4, 
                         n_jobs=-1, 
                         random_state=143, 
                         k_neighbors=k)
        X_train, y_train = resample.fit_resample(X_train, y_train)
        # fit
        model.fit(X_train, y_train)

        train_preds = model.predict(X_train)
        val_preds = model.predict(X_val)

        train_score = recall_score(y_train, train_preds)
        val_score = recall_score(y_val, val_preds)
        train_scores.append(train_score)
        val_scores.append(val_score)

    end_time = time.time()           
    res_smote[('SMOTE', k)] = {
        'ave_train_recall':np.mean(train_scores) * 100,
        'ave_val_recall':np.mean(val_scores) * 100,
        'run_time': end_time - start_time
    }

res_smote = pd.DataFrame(res_smote).T
display(res_smote)
```

```python
res_smote.ave_train_recall.plot()
res_smote.ave_val_recall.plot()
```

#### ADASYN

```python
skf = StratifiedKFold(n_splits=5)
model = GradientBoostingClassifier(random_state=11) 
res_adasyn = {}
for k in range(1, 8):
    train_scores = []
    val_scores = []
    start_time = time.time()
    for train_index, val_index in skf.split(X_trainval, y_trainval):
        X_train, X_val = X_trainval.iloc[train_index], X_trainval.iloc[val_index]
        y_train, y_val = y_trainval.iloc[train_index], y_trainval.iloc[val_index]
                    
        resample = ADASYN(sampling_strategy=0.4, 
                         n_jobs=-1, 
                         random_state=143, 
                         n_neighbors=k)
        X_train, y_train = resample.fit_resample(X_train, y_train)
        # fit
        model.fit(X_train, y_train)

        train_preds = model.predict(X_train)
        val_preds = model.predict(X_val)

        train_score = recall_score(y_train, train_preds)
        val_score = recall_score(y_val, val_preds)
        train_scores.append(train_score)
        val_scores.append(val_score)

    end_time = time.time()           
    res_adasyn[('ADASYN', k)] = {
        'ave_train_recall':np.mean(train_scores) * 100,
        'ave_val_recall':np.mean(val_scores) * 100,
        'run_time': end_time - start_time
    }

res_adasyn = pd.DataFrame(res_adasyn).T
display(res_adasyn)
```

```python
res_adasyn.ave_train_recall.plot()
res_adasyn.ave_val_recall.plot()
```

#### SMOTENC

```python
skf = StratifiedKFold(n_splits=5)
model = GradientBoostingClassifier(random_state=11) 
res_smotenc = {}
for k in range(1, 8):
    train_scores = []
    val_scores = []
    start_time = time.time()
    for train_index, val_index in skf.split(X_trainval, y_trainval):
        X_train, X_val = X_trainval.iloc[train_index], X_trainval.iloc[val_index]
        y_train, y_val = y_trainval.iloc[train_index], y_trainval.iloc[val_index]

        resample = SMOTENC(sampling_strategy=0.4, n_jobs=-1, random_state=143, 
                          categorical_features=cats, 
                          k_neighbors=k)
        X_train, y_train = resample.fit_resample(X_train, y_train)
        # fit
        model.fit(X_train, y_train)

        train_preds = model.predict(X_train)
        val_preds = model.predict(X_val)

        train_score = recall_score(y_train, train_preds)
        val_score = recall_score(y_val, val_preds)
        train_scores.append(train_score)
        val_scores.append(val_score)

    end_time = time.time()           
    res_smotenc[('SMOTENC', k)] = {
        'ave_train_recall':np.mean(train_scores) * 100,
        'ave_val_recall':np.mean(val_scores) * 100,
        'run_time': end_time - start_time
    }

res_smotenc = pd.DataFrame(res_smotenc).T
display(res_smotenc)
```

```python
res_smotenc.ave_train_recall.plot()
res_smotenc.ave_val_recall.plot()
```
